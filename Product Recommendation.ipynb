```python
customer_order_df['text_embedding_float'] = customer_order_df.text_embedding.apply(lambda x: np.array(x, dtype=float))
```

```python
customer_input = "Hi! Can you recommend a good moisturizer for me?"
```

```python
response = openai.Embedding.create(
    input=customer_input,
    model="text-embedding-ada-002"
)
embedding_customer_question = response['data'][0]['embedding']
embedding_customer_question_f = np.array(embedding_customer_question, dtype=float)
```

```python
print(type(embedding_customer_question_f))
```

```python
customer_order_df['search_purchase_history'] = customer_order_df.text_embedding.apply(lambda x: cosine_similarity(x,embedding_customer_question))
```

```python
customer_order_df.sort_values('search_purchase_history', ascending=False)
```
